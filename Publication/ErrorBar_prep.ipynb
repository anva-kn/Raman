## Thiacloprid

```python
thia_10th_percentile = np.genfromtxt('Publication/thiacloprid/thia_batch1_3min_mapping_no_cosmic_lower.csv', delimiter=',')
thia_90th_percentile = np.genfromtxt('Publication/thiacloprid/thia_batch1_3min_mapping_no_cosmic_upper.csv', delimiter=',')
```

## Acetamiprid

```python
aceta_10th_percentile = np.genfromtxt('Publication/acetamiprid/aceta_batch1_3min_mapping_no_cosmic_lower.csv', delimiter=',')
aceta_90th_percentile = np.genfromtxt('Publication/acetamiprid/aceta_batch1_3min_mapping_no_cosmic_upper.csv', delimiter=',')
```

```python
aceta_10th_percentile = aceta_10th_percentile[:,1]
aceta_90th_percentile = aceta_90th_percentile[:,1]
```

## Difference between upper and lower limit

```python
car_diff = car_90th_percentile - car_10th_percentile
thia_diff = thia_90th_percentile - thia_10th_percentile
aceta_diff = aceta_90th_percentile - aceta_10th_percentile
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(3, 1, 1)
plt.plot(f_sup, car_diff, 'r')
plt.xlabel('Raman Shift (cm$^{-1}$)')
plt.ylabel('Intensity (a.u.)')
plt.title('Carbendazim')
plt.subplot(3, 1, 2)
plt.plot(f_sup, thia_diff, 'g')
plt.xlabel('Raman Shift (cm$^{-1}$)')
plt.ylabel('Intensity (a.u.)')
plt.title('Thiacloprid')
plt.subplot(3, 1, 3)
plt.plot(f_sup, aceta_diff, 'b')
plt.xlabel('Raman Shift (cm$^{-1}$)')
plt.ylabel('Intensity (a.u.)')
plt.title('Acetamiprid')
plt.tight_layout()
plt.show()
```
