```python
thia_peaks_idx = [360, 860, 939]

# Take each peak from thia_peak_idx and find the mean at that position for all spectra in thia_batch1_no_cosmic
# Store the mean in a new array thia_batch1_no_cosmic_mean_peaks
thia_batch1_no_cosmic_mean_peaks = np.zeros((len(thia_peaks_idx), 1))
for i in range(len(thia_peaks_idx)):
    thia_batch1_no_cosmic_mean_peaks[i] = np.mean(thia_batch1_no_cosmic[:, thia_peaks_idx[i]])
```

```python
# Take each peak from thia_peak_idx and scale it to the mean value of thia_batch1_no_cosmic_mean_peaks at that position
# Store the scaling factor in a new array thia_batch1_no_cosmic_scale_peaks where each column corresponds to a peak
thia_batch1_no_cosmic_scale_peaks = np.zeros((thia_batch1_no_cosmic.shape[0], len(thia_peaks_idx)))
for i in range(len(thia_peaks_idx)):
    thia_batch1_no_cosmic_scale_peaks[:, i] = thia_batch1_no_cosmic[:, thia_peaks_idx[i]] / thia_batch1_no_cosmic_mean_peaks[i]
```

```python
# Find the mean and standard deviation of each column in thia_batch1_no_cosmic_scale_peaks
thia_batch1_no_cosmic_mean_scale_peaks = np.mean(thia_batch1_no_cosmic_scale_peaks, axis=0)
thia_batch1_no_cosmic_std_scale_peaks = np.std(thia_batch1_no_cosmic_scale_peaks, axis=0)
thia_batch1_no_cosmic_var_scale_peaks = np.var(thia_batch1_no_cosmic_scale_peaks, axis=0)
```
